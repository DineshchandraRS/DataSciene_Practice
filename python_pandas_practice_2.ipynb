```python
import pandas as pd
```

```python
movies = pd.read_csv("imdb.csv")
```

```python
movies.head()
```

```python
# Method-1

# Filtering the values using the For loop 

booleans = []

for length in movies.duration:
    if length >= 200:
      booleans.append(True)
    else:
        booleans.append(False)
```

```python
# Converting the booleans to series so that we can inset it in a Dataframe

is_long = pd.Series(booleans)
```

```python
movies[is_long]

# Filtered Dataframe Using the For loop
```

```python
# Method-2
# Filtering the values without For loop

movies[movies.duration >= 200]
```

```python
# To get the Genre and filtered duration only

# Allowing to seleect the rows and columns by label

movies.loc[movies.duration >= 200, 'genre']
```

```python
# For multiple filters Use operators like AND & OR

movies[(movies.duration >= 200) & (movies.genre == ' genre')]
```
